```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 18})
```

```python
'''
random.randint(6)0,1,2,3,4,5 중에 뽑기
np.random.randint(6) 0,1,2,3,4,5 중에 뽑기

np.random.rand(6) : 0 ~ 1의 균일 분포 표준 정규분포 난수의 matrix 생성
np.random.rand(3,2)  --> a.shape : (16,1) | *a.shape : 16,1
'''

x = 3 # True slope
a = np.arange(-2, 2, 0.55) # 데이터가 많아 질수록 true 값에 근접한다 0.55 - > 0.1
a = a.reshape(-1, 1)
b = x*a + np.random.randn(*a.shape) # Add noise
```

```python
plt.plot(a, x*a, label="True")
plt.plot(a, b, 'r+', label='Noisy data')

U, S, VT = np.linalg.svd(a, full_matrices=False)
xtilde = VT.T @ np.linalg.inv(np.diag(S)) @U.T @b # least-squere 
xtilde2 = np.linalg.pinv(a) @ b # same as xtilde
plt.plot(a, xtilde * a, 'b--', label='Regression')
plt.xlabel('a')
plt.ylabel('b')

plt.grid('--')
plt.legend()
plt.show()
```
