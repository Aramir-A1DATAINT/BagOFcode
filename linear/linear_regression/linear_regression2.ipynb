```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os 

plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size' : 18})

# Load dataset
A = np.loadtxt(os.path.join('DATA', 'hald_ingredients.csv'), delimiter=',')
b = np.loadtxt(os.path.join('DATA', 'hald_heat.csv'), delimiter=',')
# b = pd.read_csv('./data/hald_heat.csv') 비교 해보자!

# Solve Ax=b using SVD
U, S, VT = np.linalg.svd(A,full_matrices=0)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b

plt.plot(b,'k', linewidth=2, label='Heat Data') # True relationship
# plot([1, 2, 3], [1, 2, 3], 'go-', label='line 1', linewidth=2)
# plot([1, 2, 3], [1, 4, 9], 'rs', label='line 2')
# https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(A@x, '-o', color='b', linewidth=1.5, markersize=6, label='Regression')
plt.legend()
plt.show()
```
