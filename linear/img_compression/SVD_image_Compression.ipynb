```python
from matplotlib.image import imread
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
A = imread('toystory.jpeg')
```

```python
X = np.mean(A, -1) #Conver RGB to grayscale
```

```python
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False) # Economic SVD
```

```python
S = np.diag(S)
```

```python
j = 0 

for r in (5, 20, 100):
    # Contrcut approximate, image
    Xapprox = U[:, :r] @ S[:r, :r] @ VT[:r, :]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
# 100 12% of origin data 
```

```python
plt.figure(1)
plt.semilogy(np.diag(S)) # Y (y) 
plt.title('Singular Value')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S)/np.sum(np.diag(S)))) # fraction of energy!
plt.title('Singular Value: Cumulative Sum')
plt.show()
```
